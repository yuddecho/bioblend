```python
import yaml
from bioblend.aimedorig import TransMolecule

with open('config.yaml', 'r') as f:
    config = yaml.safe_load(f)

# 登录到 TransMolecule
trans_molecule = TransMolecule(config['transmolecule_url'], config['api_key'])
```

```python
# 打印用户所有历史记录信息
trans_molecule.history.info()
```

```python
# 选择指定历史记录
trans_molecule.history.select('0b900c60f93c0654')
```

```python
# 打印当前历史所有记录数据
trans_molecule.history.content()
```

```python
# 创建历史记录
trans_molecule.history.create(name="kk")
```

```python
# 上传文件夹下所有文件
# trans_molecule.dataset.upload(file_dir='data/pharma')

# 上传单个文件
trans_molecule.dataset.upload(file_path='data/pharmapepgen/test.json')
```

```python
# 获取当前历史记录所有数据信息
data = trans_molecule.dataset.get()
print(data)
```

```python
# 打印所有工具信息
trans_molecule.tool.info()
```

```python
# 获取多肽设计工具
pharm_pep_gen = trans_molecule.tool.get_tool(tool_name="PharmaPepGen")

# 获取工具输入示例
pharm_pep_gen_in = pharm_pep_gen.inputs()
print(pharm_pep_gen_in)
```

```python
# 构建参数
pharm_pep_gen_in['input_files'][0]['id'] = data['test.json'][0]['id']
pharm_pep_gen_in['num_samples'] = 10
print(pharm_pep_gen_in)
```

```python
# 运行工具
run_info = pharm_pep_gen.run(pharm_pep_gen_in)
print(run_info)
```
