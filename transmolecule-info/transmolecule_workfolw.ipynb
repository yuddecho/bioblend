```python
import yaml
from bioblend.aimedorig import TransMolecule

with open('config.yaml', 'r') as f:
    config = yaml.safe_load(f)

# 登录到 TransMolecule
trans_molecule = TransMolecule(config['transmolecule_url'], config['api_key'])
```

```python
# 创建历史记录
trans_molecule.history.create(name="cc")

# 上传文件夹下所有文件
data = trans_molecule.dataset.upload(file_dir='data/gnina')
print(data)
```

## 调用 WorkFlow

```python
workflow = trans_molecule.workflow
```

```python
# 查看 平台已有workflow 信息
workflow.info()
```

```python
# 选择支持 workflow
workflow_id = "33b43b4e7093c91f"
workflow.selct(workflow_id)
```

```python
# 查看workfolw信息
workflow_content = workflow.content()
print(workflow_content)
```

```python
# 查看输入
input_dict = workflow_content['inputs']
print(input_dict)
```

```python
# 构造输入信息
input_dict['0'] = {'src': 'hda', 'id': data['receptor.pdb'][0]['id']}
input_dict['1'] = {'src': 'hda', 'id': data['ligand.sdf'][0]['id']}
print(input_dict)
```

```python
# 运行
output = workflow.run(inputs=input_dict)
print(output)
```

## 基于 Tool 构建 Workfolw

```python
def customize_workflow(tools, receptor: str, ligand: str):
    job = []
    
    # 药效团提取工具，传入id或者name
    pharma = tools.get_tool(tool_id="pharma")
    
    # 多肽设计工具
    pharm_pep_gen = tools.get_tool(tool_name="PharmaPepGen")
    
    # 构造参数
    pharma_input = pharma.inputs()
    pharma_input['pdb_file']['id'] = receptor
    pharma_input['sdf_file']['id'] = ligand
    
    # 执行第一步
    job.append(pharma.run(pharma_input))
    
    # 构建第二步输入
    pharm_pep_gen_in = pharm_pep_gen.inputs()
    pharm_pep_gen_in['input_files'][0]['id'] = job[0]['outputs'][0]['id']
    pharm_pep_gen_in['num_samples'] = 10
    
    # 执行第二步
    job.append(pharm_pep_gen.run(pharm_pep_gen_in))
    
    return job
```

```python
# 执行
res = customize_workflow(trans_molecule.tool, receptor=data['receptor.pdb'][0]['id'], ligand=data['ligand.sdf'][0]['id'])
print(res)
```
